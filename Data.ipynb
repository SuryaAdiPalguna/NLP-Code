# 0. Library

```python
# panggil library
import nltk
import pandas as pd
```

# 1.1. Kalimat

```python
# ambil data kalimat
# df_sentences_ade = pd.read_excel("raw/trained_kalimat_ade.xlsx", usecols=["Sentence", "Parsed"])
# df_sentences_adi = pd.read_excel("raw/trained_kalimat_adi.xlsx", usecols=["Sentence", "Parsed"])
# df_sentences_krisna = pd.read_excel("raw/trained_kalimat_krisna.xlsx", usecols=["Sentence", "Parsed"])
# df_sentences = pd.concat([df_sentences_ade, df_sentences_adi, df_sentences_krisna], ignore_index=False)
df_sentences = pd.read_excel("raw/df_kalimat_fixed.xlsx", usecols=["Sentence", "Parsing"])
df_sentences = df_sentences.rename(columns={"Parsing": "Parsed"})
df_sentences
```

```python
# ekstak data parsing kalimat untuk dikumpulkan production nya
df_sentences["Sentence"] = df_sentences["Sentence"].apply(lambda x: x.lower().split())
df_sentences["Parsed"] = df_sentences["Parsed"].apply(lambda x: nltk.Tree.fromstring(x))
df_sentences["Productions"] = df_sentences["Parsed"].apply(lambda x: x.productions())
df_sentences
```

### 1.1.1. Tokenization

```python
# buat tabel sentence rules
df_sentences_rules = pd.DataFrame({"production": pd.Series(), "sum_production": pd.Series(), "sum_lhs": pd.Series()})
for productions in df_sentences["Productions"]:
  for production in productions:
    df_sentences_rules = pd.concat([df_sentences_rules, pd.DataFrame({"production": pd.Series([production]), "sum_production": pd.Series([1]), "sum_lhs": pd.Series([0])})], ignore_index=True)
df_sentences_rules["lhs"] = df_sentences_rules["production"].apply(lambda n: n.lhs())
df_sentences_rules["production"] = df_sentences_rules["production"].astype(str)
df_sentences_rules["lhs"] = df_sentences_rules["lhs"].astype(str)
df_sentences_rules
```

### 1.1.2. Normalization

```python
# buang production yang cacat
def filter_invalid_productions(prod):
  if '->' in prod:
    if prod.strip().startswith('->') or prod.strip().endswith('->'):
      return False
  return True
df_sentences_rules = df_sentences_rules[df_sentences_rules["production"].apply(filter_invalid_productions)]
df_sentences_rules
```

```python
# menghitung total masing masing production
df_sentences_rules = df_sentences_rules.groupby("production").agg({
  "lhs": "first",
  "sum_production": "sum",
  "sum_lhs": "sum"
}).reset_index()
df_sentences_rules
```

# 1.2. Kelas Kata

### 1.2.1. Case Folding

```python
# ambil data kelas kata
df_words = pd.read_excel("raw/all_kelas_kata.xlsx", usecols=["Word", "POS Tag"])
df_words["Word"] = df_words["Word"].apply(lambda w: w.lower()) # preprocessing
df_words
```

### 1.2.2. Normalization

```python
# normalisasi kelas kata
def label_kelas_kata(pos_tag):
  if pos_tag in ["NNP"]: return "PropNoun"
  if pos_tag in ["Z", "NEG", "DT", "MD", "RP"]: return "Det"
  if pos_tag in ["CC", "IN", "CS"]: return "Prep"
  if pos_tag in ["VB"]: return "Verb"
  if pos_tag in ["NN", "UH", "WH"]: return "Noun"
  if pos_tag in ["JJ"]: return "Adj"
  if pos_tag in ["PR"]: return "Pronoun"
  if pos_tag in ["CD", "NND"]: return "Num"
  if pos_tag in ["RB"]: return "Adv"
df_words["POS Tag"] = df_words["POS Tag"].apply(label_kelas_kata) # preprocessing
df_words = df_words.drop_duplicates(ignore_index=True) # preprocessing
df_words
```

```python
# buat production dari data kelas kata
df_words = df_words[["POS Tag", "Word"]].rename(columns={"POS Tag": "lhs", "Word": "rhs"})
df_words["lhs"] = df_words["lhs"].apply(lambda l: nltk.Nonterminal(l))
df_words["rhs"] = df_words["rhs"].apply(lambda r: [nltk.Nonterminal("'{}'".format(r))])
df_words["production"] = df_words.apply(lambda n: nltk.Production(lhs=n["lhs"], rhs=n["rhs"]), axis=1)
df_words
```

```python
# buat tabel df_words rules yang berisi production, lhs, sum_production, sum_lhs
df_words_rules = pd.concat([df_words.drop(columns=["rhs"])], ignore_index=True)
df_words_rules["sum_production"] = pd.Series(0, index=range(len(df_words_rules)))
df_words_rules["sum_lhs"] = pd.Series(0, index=range(len(df_words_rules)))
df_words_rules
```

# 1.3. Normal Rules

```python
# buat normal rule nya
normal_rules = pd.DataFrame({"production": [
  "K -> S P", "K -> S P Pel", "K -> S P O", "K -> S P Ket", "K -> S P O Pel", "K -> S P O Ket", "K -> S P O Pel Ket",
  "S -> NP",
  "P -> VP", "P -> NP", "P -> AdjP", "P -> NumP", 
  "O -> NP",
  "Pel -> NP", "Pel -> AdjP", "Pel -> VP", "Pel -> NumP",
  "Ket -> PP",
  "NP -> Noun", "NP -> PropNoun", "NP -> Pronoun", "NP -> NP Noun", "NP -> NP AdjP", "NP -> NP PropNoun", "NP -> NP Pronoun", "NP -> NP Det",
  "VP -> Verb", "VP -> Adv VP", "VP -> VP Verb", "VP -> VP Adj"
  "AdjP -> Adv AdjP", "AdjP -> Adj",
  "PP -> Prep NP",
  "NumP -> Num", "NumP -> NumP Num", "NumP -> NumP Noun"
]})
normal_rules["lhs"] = normal_rules["production"].apply(lambda p: nltk.CFG.fromstring(p).productions()[0].lhs())
normal_rules["sum_production"] = pd.Series(0, index=range(len(normal_rules)))
normal_rules["sum_lhs"] = pd.Series(0, index=range(len(normal_rules)))
normal_rules
```

# 2. Process

### 2.1. Olah Sentences Rules dan Normal Rules

```python
# catat data sentences rules sesuai dengan normal rules
for i, value_i in df_sentences_rules.iterrows():
  for j, value_j in normal_rules.iterrows():
    if value_i["production"] == value_j["production"]:
      normal_rules.loc[j, "sum_production"] = value_i["sum_production"]
normal_rules
```

```python
# seleksi rule yang mengandung kata-kata
df_sentences_rules = df_sentences_rules[df_sentences_rules['production'].str.contains(r"'.*'", regex=True)]
df_sentences_rules
```

### 2.2. Merge dengan Kelas Kata Rules

```python
# merge normal rules, words rule, dan sentences rules
probabilistic_rules = pd.concat([normal_rules, df_words_rules, df_sentences_rules])
# probabilistic_rules = probabilistic_rules.drop_duplicates(ignore_index=True)
probabilistic_rules
```

### 3. Hitung Probability Rule

```python
# menghitung total masing masing production
probabilistic_rules = probabilistic_rules.groupby("production").agg({
  "lhs": "first",
  "sum_production": "sum",
  "sum_lhs": "sum"
}).reset_index()
probabilistic_rules
```

```python
probabilistic_rules["sum_lhs"] = probabilistic_rules.groupby("lhs")["sum_production"].transform("sum")
probabilistic_rules
```

```python
probabilistic_rules = probabilistic_rules[probabilistic_rules["sum_lhs"] > 0]
probabilistic_rules
```

```python
probabilistic_rules['probability'] = probabilistic_rules['sum_production'] / probabilistic_rules['sum_lhs']
probabilistic_rules
```

```python
# simpan probabilistic rules
probabilistic_rules.to_csv("rules/probabilistic_rules.csv", index=False)
```
